## 1D FDTD initial conditions and boundary conditions

Start with $n = 1$, $E^1 = 0$,  $H^1 = 0$

then consider the source term $j$

$$
j_z|_i^{n} = A(\Delta t (n+1/2))e^{-2\pi i f \Delta t (n + 1/2)}j_z(t = 0)|_i
$$
here $j_z(t = 0)|_i$ is the spatial distribution of the source, $e^{-2\pi i f \Delta t (n + 1/2)}$ is the carrier, 
and $A(\Delta t (n+1/2))$ is the Envelope.

furthermore we use **PEC doundary conditions**.

## 3D FDTD: Array Sizes & Boundary Conditions

- Permittivity grid and output grid: $\text{size}(\epsilon) = [N_x,N_y,N_z]$
- Fields:
    - Tangential E-Fields and notmal H-Fields are stored at *integer indices* $\{1:N\}$
    - Normal E-Fields and tangential H-Fields are stored at *fractional indeces*, these are transformed from $\{1.5:N-0.5\rightarrow N-1\} $
- This results in the array sizes:
    - $E_x\in\{N_x-1,N_y,N_z\}$, $H_x\in\{N_x-1,N_y,N_z\}$
    - $E_y\in\{N_x,N_y-1,N_z\}$, $H_y\in\{N_x,N_y-1,N_z\}$
    - $E_z\in\{N_x,N_y,N_z-1\}$, $H_z\in\{N_x,N_y,N_z-1\}$
- PEC boundary conditions: At the boundaries the tangential E-fields and the normal H-fields are set to zero and are not updated
    - $E_x(:,1,:) = 0$, $E_x(:,N_y,:) = 0$
    - $E_x(:,:,1) = 0$, $E_x(:,:,N_z) = 0$
    - $H_x(1,:,:) = 0$, $H_x(N_x,:,:) = 0$
    - $E_y(1,:,:) = 0$, $E_y(N_x,:,:) = 0$
    - $E_y(:,:,1) = 0$, $E_y(:,:,N_z) = 0$
    - $H_y(:,1,:) = 0$, $H_y(:,N_y,:) = 0$
    - $E_z(1,:,:) = 0$, $E_z(N_x,:,:) = 0$
    - $E_z(:,1,:) = 0$, $E_z(:,N_y,:) = 0$
    - $H_z(:,:,1) = 0$, $H_z(:,:,N-z) = 0$
- Source term:
    - $j_x|_{i,j,k}^{n} = A(\Delta t (n+1/2))e^{-i \omega \Delta t (n + 1/2)}j_x(t = 0)|_{i,j,k}$
    - $j_y|_{i,j,k}^{n} = A(\Delta t (n+1/2))e^{-i \omega \Delta t (n + 1/2)}j_y(t = 0)|_{i,j,k}$
    - $j_z|_{i,j,k}^{n} = A(\Delta t (n+1/2))e^{-i \omega \Delta t (n + 1/2)}j_z(t = 0)|_{i,j,k}$
    
- Then the fields are updated as:
    - $E_x|_{i,j,K}^{n+1}$ with $i=1:N_x-1$, $j=2:N_y-1$, $k=2:N_z-1$
    - $E_y|_{i,j,K}^{n+1}$ with $i=2:N_x-1$, $j=1:N_y-1$, $k=2:N_z-1$
    - $E_z|_{i,j,k}^{n+1}$ with $i=2:N_x-1$, $j=2:N_y-1$, $k=1:N_z-1$
    - $H_x|_{i,j,k}^{n+1}$ with $i=2:N_x-1$, $j=1:N_y-1$, $k=1:N_z-1$
    - $H_y|_{i,j,k}^{n+1}$ with $i=1:N_x-1$, $j=2:N_y-1$, $k=1:N_z-1$
    - $H_z|_{i,j,k}^{n+1}$ with $i=1:N_x-1$, $j=1:N_y-1$, $k=2:N_z-1$

## Interpolation

- For postprocessing it is better to havel all fields on a common gird in space and time
- For this purpose the fields need to be interpolated correct.

therefor we need to interpolate $E_i$ over $i$ and $H_i$ over $j,k$ and $t$. 

then the interpolation fuctions are:

**Electric Field**

$$
E_{x}^{out}|_{i,j,k}^{n+1}
= \frac{1}{2}
\bigg[
E_x|_{i-1,j,k}^{n+1} + E_x|_{i,j,k}^{n+1}
\bigg]
$$

$$
E_{x}^{out}|_{i,j,k}^{n+1}
= \frac{1}{2}
\bigg[
E_x|_{i-1,j,k}^{n+1} + E_x|_{i,j,k}^{n+1}
\bigg]
$$

$$
E_{x}^{out}|_{i,j,k}^{n+1}
= \frac{1}{2}
\bigg[
E_x|_{i-1,j,k}^{n+1} + E_x|_{i,j,k}^{n+1}
\bigg]
$$


**Magnetic Field**


$$
H_{x}^{out}|_{i,j,k}^{n+1}
= \frac{1}{8}
\bigg[
H_x|_{i,j-1,k-1}^{n} + H_x|_{i,j-1,k}^{n}+H_x|_{i,j,k-1}^{n} + H_x|_{i,j,k}^{n}
+H_x|_{i,j-1,k-1}^{n+1} + H_x|_{i,j-1,k}^{n+1}+H_x|_{i,j,k-1}^{n+1} + H_x|_{i,j,k}^{n+1}
\bigg]
$$

$$
H_{y}^{out}|_{i,j,k}^{n+1}
= \frac{1}{8}
\bigg[
H_y|_{i-1,j,k-1}^{n} + H_y|_{i-1,j,k}^{n}+H_y|_{i,j,k-1}^{n} + H_y|_{i,j,k}^{n}
+H_y|_{i-1,j,k-1}^{n+1} + H_y|_{i-1,j,k}^{n+1}+H_y|_{i,j,k-1}^{n+1} + H_y|_{i,j,k}^{n+1}
\bigg]
$$


$$
H_{z}^{out}|_{i,j,k}^{n+1}
= \frac{1}{8}
\bigg[
H_z|_{i-1,j-1,k}^{n} + H_z|_{i-1,j,k}^{n}+H_z|_{i,j-1,k}^{n} + H_z|_{i,j,k}^{n}
+H_z|_{i-1,j-1,k}^{n+1} + H_z|_{i-1,j,k}^{n+1}+H_z|_{i,j-1,k}^{n+1} + H_y|_{i,j,k}^{n+1}
\bigg]
$$


### Missing values
How do we handel missing values at the boundaries?
- Interpolation of $E_{x}^{out}(1,:,:)$ requires  $E_{x}^{out}(0,:,:)$
- Interpolation of $H_{x}^{out}(:,1,:)$ requires  $H_{x}^{out}(:,1,:)$
- Interpolation of $H_{x}^{out}(:,:,1)$ requires  $H_{x}^{out}(:,:,1)$
 
At the PEC boundary the following mirror symmetries hold:
- $\textbf{E}_{||}^{-} = - \textbf{E}_{||}^{+}$, $\textbf{E}_{\perp}^{-} = \textbf{E}_{\perp}^{+}$
- $\textbf{H}_{||}^{-} =  \textbf{H}_{||}^{+}$, $\textbf{H}_{\perp}^{-} = - \textbf{H}_{\perp}^{+}$
